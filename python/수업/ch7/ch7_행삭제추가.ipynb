```python
#append() 행 추가
#dictionary, series,df + df 가능, but! dict, series로 된 데이터 추가 시, 데이터의 index명 지정불가, 오히려 기존 df의 index 사라질 수도 있음.
```

```python
#create dataframe

#array to df
print('array to df')
array1 =  np.array([[1,2,3],[5,6,7]])
array_to_df = pd.DataFrame(array1)
display(array_to_df)

#dict to df
print('dict to df')
dict1 = {'a':[1,2],'b':[3,5],'c':[8,7],'d':[2,4]}
display(pd.DataFrame(dict1))

#series to df
print('series to df')

series1 = pd.Series({'H':1,'F':4,'J':8})
display(pd.DataFrame(series1))
```

```python
# df 
print('df')
df1 = pd.DataFrame(data=[3,4,2],index=range(0,3),columns=['A'])
display(df1)
df2 = pd.DataFrame(data=[1,8,3],columns=['A'])
df3 = pd.DataFrame(data=[5,3,4],columns=['B'])
df4 = df1.append(df2,ignore_index=True)
df4_1 = df1.append(df2)

df5 = df1.append(df3,ignore_index=True)
df5_1 = df1.append(df3)

display(df4)
display(df4_1)
display(df5)
display(df5_1)
```

## 1. 다음은 두 개의 숫자를 입력받아 더하여 돌려주는 프로그램이다. 3과 6을 입력했을 때 9가 아닌 36이라는 결괏값을 돌려주었다. 이 프로그램의 오류를 수정해 보자. 

```python
input1 = int(input("첫번재 숫자를 입력하세요:"))
input2 = int(input("두번째 숫자를 입력하세요:"))

total = input1 + input2
print("두 수의 합은 %d 입니다" % total)
```

## 2. 아래와 같이 총 10줄로 이루어진 sample.txt 파일이 있다. 아래와 같이 sample.txt 파일의 숫자값을 모두 읽어 총합과 평균값을 구한 후 평균값을  result.txt 라는 파일에 쓰는 프로그램을 작성해 보자. 
### 아래에 구글 드라이브 파일 읽기에 관한 정보가 필요해 주석 부분을 주의깊게 읽어주세요 :D (local에서 jupyternotebook과 같은 파일로 실행한 분들은 필요가 없습니다.)

```python
f= open("../data/sample.txt",'r') 
lines = f.readlines()
f.close()

total = 0
for line in lines:
    score = int(line)
    total += score

average = total/len(lines)

f = open("../data/result.txt",'w')  
f.write(str(average))
f.close()
```

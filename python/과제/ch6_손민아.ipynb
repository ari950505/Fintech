```python
import numpy as np
from numpy import sin, cos
import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

## 1. sin‚Å°(ùë•+ùúã) ÏôÄ cos‚Å°(ùë•) Î•º Ìïú plotÏúÑÏóê Í∑∏Î¶¨ÏãúÏò§


Ï°∞Í±¥

- Figure Ï†úÎ™©: Signal Processing
- XÍ∞í Î≤îÏúÑ: 0~6ùúã
- XÏ∂ï Ïù¥Î¶Ñ: time
- YÏ∂ï Ïù¥Î¶Ñ: signal
- Î≤îÎ°Ä: sin‚Å°(ùë•+ùúã) Í∑∏ÎûòÌîÑÎäî ‚Äúsine‚Äù
- Î≤îÎ°Ä: ùëêùëúùë†‚Å°(ùë•) Í∑∏ÎûòÌîÑÎäî ‚Äúcosine‚Äù
- Ïù¥ Îïå Î≤îÎ°ÄÎäî Í∑∏ÎûòÌîÑÏùò Ïò§Î•∏Ï™Ω ÏÉÅÎã®Ïóê ÏúÑÏπò
- sinÍ∑∏ÎûòÌîÑÎäî Îπ®Í∞ÑÏÉâ, cosÍ∑∏ÎûòÌîÑÎäî ÌååÎûÄÏÉâÏúºÎ°ú ÌëúÌòÑ


```python
x = np.arange(0,6*np.pi,0.1)
y = np.sin(x+np.pi)
z= np.cos(x)

plt.title('Signal Processing') 
plt.plot(x,y,label='sine',color='red')
plt.plot(x,z,label='cosine',color='blue')
plt.xlabel('time')
plt.ylabel('signal')
plt.legend(loc='upper right')
plt.show()
```

## 2. sin‚Å°(ùë•+ùúã) ÏôÄ cos‚Å°(ùë•) Î•º subplotÏúºÎ°ú Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∑∏Î¶¨ÏãúÏò§.


```python
x = np.arange(0,6*np.pi,0.1)
y = np.sin(x+np.pi)
z= np.cos(x)

plt.figure(figsize=(11,6))
plt.subplot(1,2,1) , plt.plot(x,y,color='red')
plt.xlabel('time')
plt.ylabel('sin')
plt.xticks([0,10])

plt.subplot(1,2,2), plt.plot(x,z,color='blue')
plt.xlabel('time')
plt.ylabel('cos')
plt.xticks([0,10])

plt.show()
```

## 3. seoul.csv ÌååÏùºÏóêÎäî ÏÑúÏö∏Ïùò 1908ÎÖÑ 1Ïõî 1ÏùºÎ∂ÄÌÑ∞ 2020ÎÖÑ 12Ïõî 31ÏùºÏùºÍπåÏßÄÏùò ÏùºÎ≥Ñ Í∏∞Ïò® ÏûêÎ£åÍ∞Ä Ï†ÄÏû•ÎêòÏñ¥ÏûàÎã§. Ï†ÑÏ≤¥ Í∏∞Í∞Ñ ÎèôÏïà ÏµúÍ≥†Í∏∞Ïò®Ïóê ÎåÄÌï¥ÏÑú ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î¶¨ÏãúÏò§. (Í≤∞Ï∏°ÏπòÎäî Í≥µÎ∞±ÏúºÎ°ú Ï†ÄÏû•ÎêòÏñ¥ ÏûàÏùå)


```python
with open('../data/seoul.csv', 'r', encoding='cp949') as f:
    seoul = [k for k in csv.reader(f)]
```

```python
seoul = seoul[1:]
len(seoul)
```

```python
high = []
for row in seoul :
    if row[-1] !='':
        high.append(float(row[-1]))
len(high)     
plt.hist(high, bins=100)
plt.show()
```

## 4. 8ÏõîÏùò ÏµúÍ≥†Í∏∞Ïò®Îßå Ïù¥Ïö©ÌïòÏó¨ ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î¶¨ÏãúÏò§.


```python
aug = []
for row in seoul :
    if row[-1] !='':
        if row[0].split('-')[1] == '08':
            aug.append(float(row[-1]))

plt.hist(aug, bins=20)
plt.show()
```

## 5. 1ÏõîÍ≥º 8ÏõîÏùò ÏµúÍ≥†Í∏∞Ïò®Ïóê ÎåÄÌï¥ÏÑú ÏÉÅÏûêÍ∑∏Î¶ºÏùÑ Í∑∏Î¶¨ÏãúÏò§.


```python
jan = []
for row in seoul :
    if row[-1] !='':
        if row[0].split('-')[1] == '01':
            jan.append(float(row[-1]))
plt.boxplot([jan,aug])
plt.show()
```

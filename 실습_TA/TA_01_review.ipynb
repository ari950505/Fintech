### 결측치 처리

```python
# 1. age 결측치 0으로 채우기
df['age'] = df['age'].fillna(0)
df.age.isnull().sum()
```

```python
# 2. age 컬럼의 결측치 삭제
df_age = df.dropna(subset=['age']) #subset
print(len(df_age)) #len(df) : df 길이 파악
df_age.reset_index(drop=True) 
# reset_index():인덱스 재설정,drop=True 전 인덱스 열 버리기
```

```python
#age 값 float -> integer로 변환 
df.age = df.age.astype(int)
df.info()
```

```python
# pd.crosstab (교차테이블) 두 개의 변수밖에 인식 못해줌.
pd.crosstab(df.age,df.survived) #생존자 중 연령대 파악해보기 
```

```python
#margin=True -> all이라는 값의 총 갯수 컬럼 생김
pd.crosstab(df.age,df.survived,margins=True )
```

## 변수의 재정의 및 변수 생성 / 그룹화

```python
# 연속형 대출 금액 변수
# linspace(a,b,c): a에서 b 사이를 c개의 구간으로 분할
# digitize(범주형으로 변환하고 싶은 연속형 변수, d)

d = np.linspace(500,35000,5)
loan['loan_amnt_group']=np.digitize(loan['loan_amnt'],d)
loan
```

```python
loan.groupby('loan_amnt_group').describe()['loan_amnt']
```

```python
loan.groupby(['term','grade']).mean()['annual_inc']
```

```python
loan.corr()
```

```python
group_dummy_mat=pd.get_dummies(loan['loan_amnt_group'])
group_dummy_mat
```

```python
# join

group_add_dummy=loan.join(group_dummy_mat.add_prefix('loan_group_'))
group_add_dummy
```

# Merge

```python
df1 = pd.DataFrame({'key': ['b','b','a','c','a','a','b'], 'data1':range(7)})
df2 = pd.DataFrame({'key': ['a','b','d'], 'data2':['x','y','z']})
```

```python
df2
```

```python
df1
```

```python
aa = pd.merge(df1,df2)
aa
```

```python
pd.merge(df1,df2,on='key')
```

```python
df3= pd.DataFrame({'lkey': ['b','b','a','c','a','a','b'], 'data1':range(7)})
df4= pd.DataFrame({'rkey': ['a','b','d'], 'data2':range(3)})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,how='outer')
```

### 결측치 처리

````python
# 1. age 결측치 0으로 채우기
df['age'] = df['age'].fillna(0)
df.age.isnull().sum()
````

````python
# 2. age 컬럼의 결측치 삭제
df_age = df.dropna(subset=['age']) #subset
print(len(df_age)) #len(df) : df 길이 파악
df_age.reset_index(drop=True) 
# reset_index():인덱스 재설정,drop=True 전 인덱스 열 버리기
````

````python
#age 값 float -> integer로 변환 
df.age = df.age.astype(int)
df.info()
````

````python
# pd.crosstab (교차테이블) 두 개의 변수밖에 인식 못해줌.
pd.crosstab(df.age,df.survived) #생존자 중 연령대 파악해보기 
````

pd.crosstab()

````python
#margin=True -> all이라는 값의 총 갯수 컬럼 생김
pd.crosstab(df.age,df.survived,margins=True )
````

````python
#normalize=True 정규화
pd.crosstab(df.age,df.survived,normalize=True,margins=True)
````

````python
pd.crosstab([df.age,df.sex],df.survived,margins=True ) #변수3개 보는 법
````

````python
# linspace(a,b,c): a에서 b 사이를 c 간격으로 분할한다.
# digitize(연속형 변수를 범주형으로 변환하기,연속형 data -> category로 분류하기)
# 연속형 age를 0-80대 연령대로 나눠서 범주형으로 변환해보기 
d = np.linspace(0,80,9)
df['age_group'] = np.digitize(df.age,d)
df
````

merge/join

- merge(A,B)
- A.join(B) -> join은 단순합치기

```
merge 

how ='inner' (natural join)
how ='outer'(full outer join)
how ='left' (left outer join)
how ='right'(right outer join)
```



````python
df1 = pd.DataFrame({'key':['a','b','c','d','d','a'],'d1':range(6)})
df2 = pd.DataFrame({'key':['a','b','c','e'],'d2':['x','y','z','w']})
````

````python
df1
````

````python
df2
````

inner join

````python
aa = pd.merge(df1,df2) #default inner join
aa
````

````python
df1.merge(df2) #inner join
````

````python
df1.merge(df2)
````

````python
bb = pd.merge(df1, df2, on='key') #inner join
bb
````

outer join

````python
cc = pd.merge(df1,df2,how='outer')
cc
````

````python
df3 = pd.DataFrame({'leftkey':['a','b','c','d','d','a'],'d1':range(6)})
df4 = pd.DataFrame({'rightkey':['a','b','c','e'],'d2':['x','y','z','w']})
````

````python
pd.merge(df3,df4, left_on='leftkey',right_on='rightkey')
````

````python
pd.merge(df1,df2,on='key',how='left')
````

````python
pd.merge(df1,df2,on='key',how='right')
````
